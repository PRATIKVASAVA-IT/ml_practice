# 1 sample Ttest.

## Null hypothesis:
     -   variable val1 mean is same as whole feature mean.

```python

r1=data.val1[:30]
r1_mean=np.mean(r1)
print(r1_mean)
from scipy.stats import ttest_1samp
tset,pval=ttest_1samp(r1,0.025)
print("p-values",pval)
if(pval<0.05):
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

# Paired Ttest.

```python
r1=data.val1[:30]
r2=data.val2[:30]
r1_mean=np.mean(r1)
r2_mean=np.mean(r2)
print("Reading1 mean value:",r1_mean)
print("Reading2 mean value:",r2_mean)
r1_std=np.std(r1)
r2_std=np.std(r2)
print("Reading1 std value:",r1_std)
print("Reading2 std value:",r2_std)
ttest,pval=ttest_ind(r1,r2)
print("p-value",pval)
if pval <0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```

# Stests

```python
from scipy import stats
from statsmodels.stats import weightstats as stests
ztest,pval=stests.ztest(data['val1'],x2=None,value=0.031)
print(float(pval))
if pval<0.05:
    print("We reject null hypothesis")
else:
    print("We accept null hypothesis")
```

# One way Annova

```python
from scipy.stats import f_oneway
F,p=f_oneway(data.val1,data.val2)
print("p-value for significance is:",p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

# Chi square test

```python
contingency_table=pd.crosstab(data["val1"],data["val2"])
print('contingency_table :-\n',contingency_table)
Observed_Values=contingency_table.values 
print("Observed Values :-\n",Observed_Values)
b=stats.chi2_contingency(contingency_table)
Expected_Values=b[3]
print("Expected Values :-\n",Expected_Values)
no_of_rows=len(contingency_table.iloc[0:2,0])
no_of_columns=len(contingency_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
```

```python
alpha = 0.05
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
print("chi-square statistic:-",chi_square_statistic)
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('chi-square statistic:',chi_square_statistic)
print('critical_value:',critical_value)
print('p-value:',p_value)
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

# Making Data appropriate for train and test.

```python
X=data.drop('result',axis=1) 
y=data['result']
```

# Splitting Data.

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0) 
```

# Fitting Data.

```python
from sklearn.tree import DecisionTreeClassifier 
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train) 
dtree_predictions = dtree_model.predict(X_test) 
dtree_predictions
```

```python
from sklearn.metrics import confusion_matrix 
cm=confusion_matrix(y_test,dtree_predictions)
cm
```
